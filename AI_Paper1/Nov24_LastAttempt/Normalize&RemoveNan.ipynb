### Remove NaN values from ellipse arrays and across... all others

```python
#relevant import statements
import numpy as np
import math
import xarray as xr 
import pickle 
import matplotlib.pyplot as plt
```

```python
infile = open("./Updated_Diags/two_wind_1959.p", 'rb') 
wind10 = pickle.load(infile)
infile.close()

infile = open("./Updated_Diags/two_size_1959.p", 'rb') 
size10 = pickle.load(infile)
infile.close()

infile = open("./Updated_Diags/two_cenlat_1959.p", 'rb')
cenlat10 = pickle.load(infile)
infile.close()

infile = open("nao_vals.p", 'rb')
nao = pickle.load(infile)
infile.close()
```

```python
#eddy-heat flux anonym MDTF at 100hPa 
infile = open("./Updated_Diags/1959ZMeddyheatflux.p", 'rb') 
ehf100 = pickle.load(infile)
infile.close()
```

```python
wind10.shape
```

```python
#remove repeat year
wind10 = np.delete(wind10,[20],0)
cenlat10 = np.delete(cenlat10,[20],0)
size10 = np.delete(size10,[20],0)
```

```python
print(ehf100.shape)
```

```python
ehf100 = np.delete(ehf100,[120],1)
```

```python
wind10.shape
```

```python
nao = nao.reshape((63,151))
```

```python
clt = (cenlat10 - np.nanmean(cenlat10)) / (np.nanstd(cenlat10))
clt
```

```python
sz = (size10 - np.nanmean(size10)) / (np.nanstd(size10))
sz
```

```python
wd = (wind10 - np.nanmin(wind10)) / (np.nanmax(wind10) - np.nanmin(wind10))
wd
```

```python
ehf = (ehf100 - np.nanmean(ehf100)) / (np.nanstd(ehf100))
ehf
```

```python
nao = (nao - np.nanmean(nao)) / (np.nanstd(nao))
nao
```

```python
indices = np.isnan(wd)
for i in range(0,63):
    for j in range(0,151):
        if indices[i,j] != False:
            print(i)
            print(j)
            print("True")
            wd[i,j] = 0
            clt[i,j] = 0
            sz[i,j] = 0
            ehf[i,j] = 0
            nao[i,j] = 0
```

```python
##export out NaN removed variables
pickle.dump(sz, open("1959size10_NaN.p", 'wb')) 
pickle.dump(clt, open("1959cenlat10_NaN.p", 'wb'))
pickle.dump(ehf, open("1959ehf100_NaN.p", 'wb')) 
pickle.dump(wd, open("1959wind10_NaN.p", 'wb'))
pickle.dump(nao, open("1959nao_NaN.p", 'wb'))
```
