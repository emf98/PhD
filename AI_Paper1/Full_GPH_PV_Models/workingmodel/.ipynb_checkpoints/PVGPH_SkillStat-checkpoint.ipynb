### Model architecture specifics start here.

```python
###### batch, epochs, etc
batch_size = 256 #The number of samples the network sees before it backpropagates (batch size)
epochs = 50 #The number of times the network will loop through the entire dataset (epochs)
shuffle = True #Set whether to shuffle the training data so the model doesn't see it sequentially 
verbose = 2 #Set whether the model will output information when trained (0 = no output; 2 = output accuracy every epoch)
```

```python
##class weight creator for the instance where now I have a four dimensional output array 
def class_weight_creator(Y):
    class_dict = {}
    Y_reshaped = Y.reshape(-1, Y.shape[-1])
    weights = np.max(np.sum(Y_reshaped, axis=0)) / np.sum(Y_reshaped, axis=0)
    for i in range(Y.shape[-1] ):
        class_dict[i] = weights[i]
        
    return class_dict
```

```python
def combined(number_inputs1, number_inputs2, num_cats, n1, n3, n5, reg1, reg3, reg5, learn):
    ##first layers for PVU ANN
    input_tensor1 = Input(shape=(int(1595),), name='layer1')
    layer1 = layers.Dense(n1, activation='relu', use_bias=True,
                          kernel_initializer='he_normal',bias_initializer='he_normal',
                          kernel_regularizer=keras.regularizers.l2(l2= reg1))(input_tensor1)
   
    #layer1 = layers.Dense(n2, activation='leaky_relu', use_bias=True,
                          #kernel_initializer='he_normal',bias_initializer='he_normal',
                          #kernel_regularizer=keras.regularizers.l2(l2= reg2))(input_tensor1)
    

    ## second layers for GPH ANN
    input_tensor2 = Input(shape=(int(777),), name='layer2')
    layer2 = layers.Dense(n3, activation='relu', use_bias=True,
                          kernel_initializer='he_normal',bias_initializer='he_normal',
                          kernel_regularizer=keras.regularizers.l2(l2= reg3))(input_tensor2)
    #layer2 = layers.Dense(n4, activation='leaky_relu', use_bias=True,
                          #kernel_initializer='he_normal',bias_initializer='he_normal',
                          #kernel_regularizer=keras.regularizers.l2(l2= reg4))(input_tensor2)
    ##return_sequence = False
    
    
    ##combine two model inputs
    merge_layer = Concatenate()([layer1,layer2])

    ##more dense layers
    merge = layers.Dense(n5, activation='relu', use_bias=True,
                          kernel_initializer='he_normal',bias_initializer='he_normal',
                          kernel_regularizer=keras.regularizers.l2(l2= reg5))(merge_layer)

    #decay_rate = learn / epochs
    output_tensor = layers.Dense(num_cats,activation='softmax')(merge)
    model = Model([input_tensor1, input_tensor2], output_tensor)
    model.compile(optimizer=keras.optimizers.legacy.Adam(learning_rate=learn),
              loss = 'categorical_crossentropy', #Our loss function is based on categorical error
              metrics=[keras.metrics.categorical_accuracy], #We will print out the categorical accuracy as the network is trained
              )
    return model
```

```python
##definition statement for ACC
def calculate_accuracy(y_true, y_prob, threshold=0.5):
    y_pred = (y_prob >= threshold).astype(int)
    return np.mean(y_true == y_pred)
```

```python
##Early Stopping Condition, if desired
early_stopping = EarlyStopping(monitor='val_loss',patience=4,restore_best_weights=True,verbose=1)
```

```python
##full dataset climo probabilities 
climo_full = climo(Y_all,63,lag)
#climo_train = climo(Y_tri,59,lag)
#climo_test = climo(Y_tes,4,lag)
```

```python
##create loop for cross validation
for l in range(0,100):
    ##randomly choose a fraction of events for validation and training
    start = random.randrange(len(X_tri1[:,0])-frac_ind)
    end = start+(frac_ind)

    #choose random 6 year window of training data to be used for validation
    X_val1 = X_tri1[start:end]
    X_val2 = X_tri2[start:end]
    
    Y_val = Y_tri[start:end]

    #remaining 53 years of training data
    X_train1a = X_tri1[0:start]
    X_train1b = X_tri1[end:]

    X_train2a = X_tri2[0:start]
    X_train2b = X_tri2[end:]

    Y_train1 = Y_tri[0:start]
    Y_train2 = Y_tri[end:]

    ##concatenate all of these
    X_tr1 = np.concatenate((X_train1a,X_train1b))
    X_tr2 = np.concatenate((X_train2a,X_train2b))
    Y_train = np.concatenate((Y_train1,Y_train2))

    ##dictionaries for training and validation
    X_train = {'layer1': X_tr1, 'layer2': X_tr2}
    X_all = {'layer1': X_all1, 'layer2': X_all2}
    X_test = {'layer1': X_tes1, 'layer2': X_tes2}
    X_val = {'layer1': X_val1, 'layer2': X_val2}

    climo_tr1 = climo_full[0:start,:]
    climo_tr2 = climo_full[end:train_edge,:]

    climo_tr = np.concatenate((climo_tr1,climo_tr2))
    
    ##calculate class weights on remaining training data
    class_weight = class_weight_creator(Y_train)

    ##pause, calculate climo probabilities
    climo_val = climo_full[start:end,:]
    climo_test = climo_full[train_edge:,:]

    ##initalize model
    model = combined(input_shape1, input_shape2, 2, 128, 124, 28, 0.54, 0.102, 0.087, 0.008)
    ##FIT the model

    model.fit(X_train, Y_train, validation_data=(X_val, Y_val),
              batch_size=batch_size, epochs=epochs, shuffle=shuffle,
              class_weight = class_weight, callbacks=[early_stopping])
    
    train_loss.append(model.history.history['loss'])
    val_loss.append(model.history.history['val_loss'])
    
    cat_acc.append(model.history.history['categorical_accuracy'])
    val_acc.append(model.history.history['val_categorical_accuracy'])
     
    pred = model.predict(X_all)
    pred_val = model.predict(X_val)
    pred_train = model.predict(X_train)
    pred_test = model.predict(X_test)
    ########################################################################################################
    ##STATISTICS CALCULATINNG
    pred_class = []
    predval_class = []
    predtr_class = []
    predtest_class = []
    ##BRIER SKILL SCORE
    BSS(l, start, end,
        climo_full, pred, Y_all, BSS_all,
        climo_val, pred_val, Y_val, BSS_val,
        climo_tr, pred_train, Y_train, BSS_train,
        climo_test, pred_test, Y_tes, BSS_test)
    ##RECALL ACCURACY SCORE    
    RAS_two(l, Rec_all, climo_full, Y_all, pred, pred_class,
            climo_val, Rec_val, Y_val, pred_val, predval_class,
            climo_tr, Rec_train, Y_train, pred_train, predtr_class,
            climo_test, Rec_test, Y_tes, pred_test, predtest_class)
    ##PRECISION ACCURACY SCORE     
    PAS_two(l, Prec_all, climo_full, Y_all, pred, pred_class,
            climo_val, Prec_val, Y_val, pred_val, predval_class,
            climo_tr, Prec_train, Y_train, pred_train, predtr_class,
            climo_test, Prec_test, Y_tes, pred_test, predtest_class)
    ########################################################################################################
    ##ACC
    ##TESTING
    q90 = np.percentile(pred_test,90,axis=0) ##90th percentile of test
    ##90th percentile acc
    great90 = [i for i, row in enumerate(pred_test) if (row[0] > q90[0]) or (row[1] > q90[1])]
    # Create the arrays of probabilities and actual values that exceed the 90th percentile
    test90 = pred_test[great90]
    test90_norm = Y_tes[great90]
    test90_acc.append(calculate_accuracy(test90_norm, test90, threshold=0.5))
    ##full model
    fulltest_acc.append(calculate_accuracy(Y_tes, pred_test, threshold=0.5))

    ##TRAINING
    q90 = np.percentile(pred_train,90,axis=0) ##90th percentile of test
    ##90th percentile acc
    great90 = [i for i, row in enumerate(pred_train) if (row[0] > q90[0]) or (row[1] > q90[1])]
    # Create the arrays of probabilities and actual values that exceed the 90th percentile
    train90 = pred_train[great90]
    train90_norm = Y_train[great90]
    train90_acc.append(calculate_accuracy(train90_norm, train90, threshold=0.5))
    ##full model
    fulltrain_acc.append(calculate_accuracy(Y_train, pred_train, threshold=0.5))

    ##VALIDATION
    q90 = np.percentile(pred_val,90,axis=0) ##90th percentile of test
    ##90th percentile acc
    great90 = great90 = [i for i, row in enumerate(pred_val) if (row[0] > q90[0]) or (row[1] > q90[1])]
    # Create the arrays of probabilities and actual values that exceed the 90th percentile
    val90 = pred_val[great90]
    val90_norm = Y_val[great90]
    val90_acc.append(calculate_accuracy(val90_norm, val90, threshold=0.5))
    ##full model
    fullval_acc.append(calculate_accuracy(Y_val, pred_val, threshold=0.5))

    
    ########################################################################################################
    ##LRP
    # Creating an analyzer
    model_wo_sm = innvestigate.model_wo_softmax(model)
    gradient_analyzer = innvestigate.create_analyzer("deep_taylor", model_wo_sm)

    # Applying the analyzer
    analysis = gradient_analyzer.analyze(X_test)
    tf.keras.backend.clear_session()

    pv_analysis = analysis[0]
    gph_analysis =analysis[1]

    #Convert the list to an array to make processing simpler and reshape it into a 2-D map (our inputs were vectorized)
    #LRP_heatmaps = analysis.reshape(analysis.shape[0], 21, 37)

    #Normalize each heatmap to have a maximum value of 1 so all samples can be compared equally
    #LRP_heatmaps_scaled = LRP_heatmaps/np.nanmax(LRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]

    #Now take the mean across all samples to make a composite...
    #LRP_heatmaps_mean = np.nanmean(LRP_heatmaps_scaled, axis=0)

    #LRP[l,:,:] = LRP_heatmaps_mean[:,:]

    ##I am going to transition right here into making separate LRP for the top 10% most confident and correct
    #predictions in the testing data 
    q90 = np.percentile(pred_test,90,axis=0) ##identify the threshold for the 90th percentile of confidence
    ##AS before with checking the threshold
    great90 = [i for i, row in enumerate(pred_test) if (row[0] > q90[0]) or (row[1] > q90[1])]
    
    ##classify the accuracy of predicitons
    correct_pos = [] #correct positive anomaly
    correct_neg = [] #correct negative anomaly
    
    false_pos = [] #false positive
    false_neg = [] #false negative
    
    for j in range(len(great90)):
        #print(j)
        #print(great90[j])
        index = great90[j]
        if pred_test[index,0] < pred_test[index,1] and Y_tes[index,0] == 0:
            correct_pos.append(index)
            #print('###########')
        elif pred_test[index,0] > pred_test[index,1] and Y_tes[index,0] == 1:
            correct_neg.append(index)
            #print('###########')
        elif pred_test[index,0] < pred_test[index,1] and Y_tes[index,0] == 1:
            false_neg.append(index)
            #print('###########')
        elif pred_test[index,0] > pred_test[index,1] and Y_tes[index,0] == 0:
            false_pos.append(index)
            #print('###########')
            
    # Creating an analyzer
    ##apply the analyzer to testing data
    ##correct positive
    PV_posXtest = np.array(pv_analysis[correct_pos])
    PV_comp_pos[l,:] = np.nanmean(X_tes1[correct_pos],axis=0) #save for composite

    GPH_posXtest = np.array(gph_analysis[correct_pos])
    GPH_comp_pos[l,:] = np.nanmean(X_tes2[correct_pos],axis=0) #save for composite
   
    #correct negative
    PV_negXtest = np.array(pv_analysis[correct_neg])
    PV_comp_neg[l,:] = np.nanmean(X_tes1[correct_neg],axis=0) #save for composite

    GPH_negXtest = np.array(gph_analysis[correct_neg])
    GPH_comp_neg[l,:] = np.nanmean(X_tes2[correct_neg],axis=0) #save for composite
    
    #false positive
    PV_FposXtest = np.array(pv_analysis[false_pos])
    PV_comp_Fpos[l,:] = np.nanmean(X_tes1[false_pos],axis=0) #save for composite

    GPH_FposXtest = np.array(gph_analysis[false_pos])
    GPH_comp_Fpos[l,:] = np.nanmean(X_tes2[false_pos],axis=0) #save for composite
    
    #false negative
    PV_FnegXtest = np.array(pv_analysis[false_neg])
    PV_comp_Fneg[l,:] = np.nanmean(X_tes1[false_neg],axis=0) #save for composite

    GPH_FnegXtest = np.array(gph_analysis[false_neg])
    GPH_comp_Fneg[l,:] = np.nanmean(X_tes2[false_neg],axis=0) #save for composite
    
    ###make heatmap
    ##CORRECT PLOTS
    #Convert the list to an array to make processing simpler and reshape it into a 2-D map (our inputs were vectorized)
    PV_posLRP_heatmaps = PV_posXtest.reshape(PV_posXtest.shape[0], 11, 145)
    PV_posLRP_heatmaps_scaled = PV_posLRP_heatmaps/np.nanmax(PV_posLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    PV_posLRP_heatmaps_mean = np.nanmean(PV_posLRP_heatmaps_scaled, axis=0)
    PV_pos90_LRP[l,:,:] = PV_posLRP_heatmaps_mean[:,:]

    GPH_posLRP_heatmaps = GPH_posXtest.reshape(GPH_posXtest.shape[0], 21, 37)
    GPH_posLRP_heatmaps_scaled = GPH_posLRP_heatmaps/np.nanmax(GPH_posLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    GPH_posLRP_heatmaps_mean = np.nanmean(GPH_posLRP_heatmaps_scaled, axis=0)
    GPH_pos90_LRP[l,:,:] = GPH_posLRP_heatmaps_mean[:,:]
    
    PV_negLRP_heatmaps = PV_negXtest.reshape(PV_negXtest.shape[0], 11, 145)
    PV_negLRP_heatmaps_scaled = PV_negLRP_heatmaps/np.nanmax(PV_negLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    PV_negLRP_heatmaps_mean = np.nanmean(PV_negLRP_heatmaps_scaled, axis=0)
    PV_neg90_LRP[l,:,:] = PV_negLRP_heatmaps_mean[:,:]

    GPH_negLRP_heatmaps = GPH_negXtest.reshape(GPH_negXtest.shape[0], 21, 37)
    GPH_negLRP_heatmaps_scaled = GPH_negLRP_heatmaps/np.nanmax(GPH_negLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    GPH_negLRP_heatmaps_mean = np.nanmean(GPH_negLRP_heatmaps_scaled, axis=0)
    GPH_neg90_LRP[l,:,:] = GPH_negLRP_heatmaps_mean[:,:]

    ###FALSE PLOTS
    PV_FposLRP_heatmaps = PV_FposXtest.reshape(PV_FposXtest.shape[0], 11, 145)
    PV_FposLRP_heatmaps_scaled = PV_FposLRP_heatmaps/np.nanmax(PV_FposLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    PV_FposLRP_heatmaps_mean = np.nanmean(PV_FposLRP_heatmaps_scaled, axis=0)
    PV_Fpos90_LRP[l,:,:] = PV_FposLRP_heatmaps_mean[:,:]

    GPH_FposLRP_heatmaps = GPH_FposXtest.reshape(GPH_FposXtest.shape[0], 21, 37)
    GPH_FposLRP_heatmaps_scaled = GPH_FposLRP_heatmaps/np.nanmax(GPH_FposLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    GPH_FposLRP_heatmaps_mean = np.nanmean(GPH_FposLRP_heatmaps_scaled, axis=0)
    GPH_Fpos90_LRP[l,:,:] = GPH_FposLRP_heatmaps_mean[:,:]
    
    PV_FnegLRP_heatmaps = PV_FnegXtest.reshape(PV_FnegXtest.shape[0], 11, 145)
    PV_FnegLRP_heatmaps_scaled = PV_FnegLRP_heatmaps/np.nanmax(PV_FnegLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    PV_FnegLRP_heatmaps_mean = np.nanmean(PV_FnegLRP_heatmaps_scaled, axis=0)
    PV_Fneg90_LRP[l,:,:] = PV_FnegLRP_heatmaps_mean[:,:]

    GPH_FnegLRP_heatmaps = GPH_FnegXtest.reshape(GPH_FnegXtest.shape[0], 21, 37)
    GPH_FnegLRP_heatmaps_scaled = GPH_FnegLRP_heatmaps/np.nanmax(GPH_FnegLRP_heatmaps, axis=(-2,-1))[:,np.newaxis,np.newaxis]
    GPH_FnegLRP_heatmaps_mean = np.nanmean(GPH_FnegLRP_heatmaps_scaled, axis=0)
    GPH_Fneg90_LRP[l,:,:] = GPH_FnegLRP_heatmaps_mean[:,:]
```
