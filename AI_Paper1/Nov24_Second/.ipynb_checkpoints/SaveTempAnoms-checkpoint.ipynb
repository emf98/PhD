This code will be used for designating out temperature regions.

In this first instance, I am choosing an area from Europe that is known to be sensitive to the influence of the NAO per Butler et al 2017. 

```python
#import statements... I think these are all of the relevant ones to what I am doing here. 
import datetime as dt
import numpy as np
import xarray as xr
import pickle
```

```python
##I have to call back to where this data was loaded previously because it is massive. 
infile = open("../../../REDO_ProcessERA5/FullNH/FullNH_1000_temp_1959_PT1.p", 'rb') 
file1 = pickle.load(infile)
#test = pickle.load(infile)
infile.close()

infile = open("../../../REDO_ProcessERA5/FullNH/FullNH_1000_temp_1959_PT2.p", 'rb') 
file2 = pickle.load(infile)
infile.close()
```

```python
##save year index
year = [i for i in range(1959,2022)] #indicate years for data, needed for opening files
```

```python
##save lat lon indices
lats = []
lons = []
testfile = xr.open_dataset("../../../era5/gph/era5_gph_1979.nc")
lats.append(testfile.latitude.values)
lons.append(testfile.longitude.values)

lats = lats[0]
lons = lons[0]
```

```python
##save day indices
import pandas as pd

s = pd.date_range('2011-11-01', periods=152, freq='D').to_series()
test1 = s.dt.strftime('%m-%d')
times = []
for i in range(len(s)):
    times.append(test1[i])
```

```python
temp = np.concatenate((file1,file2), axis=0)
temp.shape
```

```python
##create array for gph ... 
t = xr.DataArray(data= temp, dims = ["year","day","lat","lon"],coords = dict(year = year,
                                                                                day = times,
                                                                                lat = lats,
                                                                                lon = lons))
t
```

```python
#select cap
eur_t = t.loc[dict(lat=slice(75,60), lon=slice(10,45))]
eur_t
```

```python
##pickle out these files
pickle.dump(eur_t.values, open("1959europe_temp1000.p", 'wb'))
```

```python
#create weights
weights=np.cos(np.deg2rad(eur_t.lat))
e_t_weight = eur_t.weighted(weights).mean(dim="lon").mean(dim="lat") #average over the full area
e_t_weight
```

```python
##europe
dailymean_e = np.mean(eur_t,axis=0)
anom_e = eur_t - dailymean_e
anom_e.shape
```

```python
pickle.dump(anom_e.values, open("FULLeur1000_tempANOMs.p", 'wb'))
```

```python
pickle.dump(lats, open("era5_latindex.p", 'wb'))
pickle.dump(lons, open("era5_lonindex.p", 'wb'))
pickle.dump(times, open("era5_timeindex.p", 'wb'))
```
