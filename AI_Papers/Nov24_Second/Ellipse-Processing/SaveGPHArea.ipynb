```python
#import statements... I think these are all of the relevant ones to what I am doing here. 
import datetime as dt
import numpy as np
import xarray as xr
import pickle
```

```python
infile = open("../1959_NAOgph500_2div.p", 'rb') 
gph = pickle.load(infile)
infile.close()
```

```python
##save year index
year = [i for i in range(1959,2022)] #indicate years for data, needed for opening files
```

```python
##save lat lon indices
lats = []
lons = []
testfile = xr.open_dataset("../../../../era5/gph/era5_gph_1979.nc")
lats.append(testfile.latitude.values)
lons.append(testfile.longitude.values)

lats = lats[0]
lons = lons[0]
```

```python
print(lats[40:281])
print(lons[320:681])
```

```python
##save day indices
import pandas as pd

s = pd.date_range('2011-11-01', periods=152, freq='D').to_series()
test1 = s.dt.strftime('%m-%d')
times = []
for i in range(len(s)):
    times.append(test1[i])
```

```python
##create array for gph ... 
g = xr.DataArray(data= gph, dims = ["year","day","lat","lon"],coords = dict(year = year,
                                                                                day = times,
                                                                                lat = lats[40:281],
                                                                                lon = lons[320:681]))
g
```

```python
##doesnt need to be divided by 10
```

```python
#select cap (helsinki 60,25)
g_NAO = g.loc[dict(lat=slice(70,60), lon=slice(-40,-20))]
g_NAO
```

```python
##pickle out this files
pickle.dump(g_NAO.values, open("1959NAO_gph500.p", 'wb'))
```

```python
#create weights
weights=np.cos(np.deg2rad(g_NAO.lat))
g_NAO_weight = g_NAO.weighted(weights).mean(dim="lon").mean(dim="lat") #average over the full area
g_NAO_weight
```

```python
blorp = g_NAO_weight.values
```

```python
blorp = np.delete(blorp,[120],1)
blorp.shape
```
