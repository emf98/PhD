### V and T at 100hPa 

```python
#import statements... I think these are all of the relevant ones to what I am doing here. 
import datetime as dt
import numpy as np
import xarray as xr
import pickle
```

```python
##Import Temps
infile = open("Can100_t_1959_PT1.p", 'rb') 
file1 = pickle.load(infile)
infile.close()

infile = open("Can100_t_1959_PT2.p", 'rb') 
file2 = pickle.load(infile)
infile.close()
```

```python
##save year index
year = [i for i in range(1959,2022)] #indicate years for data, needed for opening files
len(year)
```

```python
##save lat lon indices
lats = []
lons = []
testfile = xr.open_dataset("../../../../era5/gph/era5_gph_1979.nc")
lats.append(testfile.latitude.values)
lons.append(testfile.longitude.values)

lats = lats[0]
lons = lons[0]
```

```python
lat = lats[60:181]
#print(lat)
```

```python
lon = lons[240:541]
```

```python
##save day indices
import pandas as pd

s = pd.date_range('2011-11-01', periods=152, freq='D').to_series()
test1 = s.dt.strftime('%m-%d')
times = []
for i in range(len(s)):
    times.append(test1[i])

#times
```

```python
test = np.concatenate((file1,file2), axis=0)
test.shape
```

```python
##create array for temp 
t = xr.DataArray(data= test, dims = ["year","day","lat","lon"],coords = dict(year = year,
                                                                                day = times,
                                                                                lat = lat,
                                                                                lon = lon))
t
```

```python
t = t-273.15
t
```

```python
##V-Wind
infile = open("Can100_v_1959_PT1.p", 'rb') 
file1 = pickle.load(infile)
infile.close()

infile = open("Can100_v_1959_PT2.p", 'rb') 
file2 = pickle.load(infile)
infile.close()
```

```python
test = np.concatenate((file1,file2), axis=0)
test.shape
```

```python
##create array for v wind 
v = xr.DataArray(data= test, dims = ["year","day","lat","lon"],coords = dict(year = year,
                                                                                day = times,
                                                                                lat = lat,
                                                                                lon = lon))
v
```

```python
#create weights
t_weights=np.cos(np.deg2rad(t.lat))
zm_t = t.weighted(t_weights).mean(dim="lon")#average over the full area
zm_t ##zonal mean temp
```

```python
#create weights
v_weights=np.cos(np.deg2rad(v.lat))
zm_v = v.weighted(v_weights).mean(dim="lon")#average over the full area
zm_v ##zonal mean meridional
```

```python
ehf = ((v - zm_v) * (t - zm_t))
```

```python
ehf_weights=np.cos(np.deg2rad(ehf.lat))
zm_ehf = ehf.weighted(ehf_weights).mean(dim="lon").mean(dim='lat')#average over the full area
zm_ehf ##zonal mean meridional eddy heat flux
```

```python
zm_ehf
```

```python
zm_ehf = np.delete(zm_ehf,[120],1)
zm_ehf.shape
```
