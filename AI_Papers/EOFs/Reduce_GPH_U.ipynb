```python
import numpy as np
import math
import xarray as xr 
import pickle 
import matplotlib.pyplot as plt
```

```python
infile = open("./large_area_data/1959_gph_2deg.p", 'rb') 
gph = pickle.load(infile)
infile.close()

infile = open("./large_area_data/1959_u_2deg.p", 'rb') 
uwind = pickle.load(infile)
infile.close()
```

```python
lon = np.arange(0,360,2)

lat = np.arange(90,18,-2)

time = np.arange(0,183,1)

lev = np.array([100., 200., 300., 500., 700., 1000., 2000., 3000., 5000., 7000., 10000., 12500., 15000., 17500., 20000., 22500., 25000., 30000., 35000., 40000., 
                45000., 50000., 55000., 60000., 65000., 70000., 75000., 77500., 80000., 82500., 85000., 87500., 90000., 92500., 95000., 97500., 100000.])
```

```python
##save year index
year = [i for i in range(1959,2021)] #indicate years for data, needed for opening files
```

```python
gph = xr.DataArray(data= gph[:62], dims = ["year","day","lev","lat","lon"],coords = dict(year = year,
                                                                                day = time,
                                                                                lev = lev,
                                                                                lat = lat,
                                                                                lon = lon))

uwind = xr.DataArray(data= uwind[:62], dims = ["year","day","lev","lat","lon"],coords = dict(year = year,
                                                                                day = time,
                                                                                lev = lev,
                                                                                lat = lat,
                                                                                lon = lon))
```

```python
g = gph.sel(lat = slice(75,45))
g
```

```python
#create weights
g_weights=np.cos(np.deg2rad(g.lat))
zm_g = g.weighted(g_weights).mean(dim="lat")#average over the full area
zm_g ##meridional avg
```

```python
#create weights
u_weights=np.cos(np.deg2rad(uwind.lat))
zm_u = uwind.weighted(u_weights).mean(dim="lon")#average over the full area
zm_u ##zonal avg
```

```python
pickle.dump(zm_g.values, open("./processed_vertical_data/1959gph.p", 'wb'))
pickle.dump(zm_u.values, open("./processed_vertical_data/1959u.p", 'wb'))
```
