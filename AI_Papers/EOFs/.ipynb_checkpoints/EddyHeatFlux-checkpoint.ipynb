Calculate eddy heat flux. 

```python
import numpy as np
import math
import xarray as xr 
import pickle 
import matplotlib.pyplot as plt
```

```python
infile = open("./large_area_data/1959_v_2deg.p", 'rb') 
v100 = pickle.load(infile)
infile.close()

infile = open("./large_area_data/1959_t_2deg.p", 'rb') 
t100 = pickle.load(infile)
infile.close()
```

```python
t100.shape
```

```python
lon = np.arange(0,360,2)

lat = np.arange(75,45,-2)

time = np.arange(0,183,1)

lev = np.array([100., 200., 300., 500., 700., 1000., 2000., 3000., 5000., 7000., 10000., 12500., 15000., 17500., 20000., 22500., 25000., 30000., 35000., 40000., 
                45000., 50000., 55000., 60000., 65000., 70000., 75000., 77500., 80000., 82500., 85000., 87500., 90000., 92500., 95000., 97500., 100000.])
```

```python
##save year index
year = [i for i in range(1959,2021)] #indicate years for data, needed for opening files
```

```python
print(len(lon))
print(len(lat))
print(len(time))
print(len(lev))
```

```python
t100 = xr.DataArray(data= t100[:62], dims = ["year","day","lev","lat","lon"],coords = dict(year = year,
                                                                                day = time,
                                                                                lev = lev,
                                                                                lat = lat,
                                                                                lon = lon))
```

```python
v100 = xr.DataArray(data= v100[:62], dims = ["year","day","lev","lat","lon"],coords = dict(year = year,
                                                                                day = time,
                                                                                lev = lev,
                                                                                lat = lat,
                                                                                lon = lon))
```

```python
v100
```

```python
#create weights
t_weights=np.cos(np.deg2rad(t100.lat))
zm_t = t100.weighted(t_weights).mean(dim="lon")#average over the full area
zm_t ##zonal mean temp
```

```python
#create weights
v_weights=np.cos(np.deg2rad(v100.lat))
zm_v = v100.weighted(v_weights).mean(dim="lon")#average over the full area
zm_v ##zonal mean meridional
```

```python
ehf = ((v100 - zm_v) * (t100 - zm_t))
```

```python
ehf_weights=np.cos(np.deg2rad(ehf.lat))
zm_ehf = ehf.weighted(ehf_weights).mean(dim='lat')#average over the latitudinal
zm_ehf ##mean meridional eddy heat flux
```

```python
pickle.dump(zm_ehf.values, open("./processed_vertical_data/1959ZMeddyheatflux.p", 'wb'))
```
